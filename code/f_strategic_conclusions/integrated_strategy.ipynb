## Executive Summary

After analyzing two years of household shopping behavior across five interconnected workstreams, we have a clear, data driven answer to the central business challenge: how to allocate marketing resources efficiently when Premium Loyalists generate 76% of revenue while the broader base remains underpenetrated.

The integrated strategy rests on four pillars.

First, segment-first resource allocation. We need to concentrate retention and upsell efforts on the 392 Premium Loyalist households in Cluster 2 who drive 47% of revenue, while deploying targeted reactivation for the at risk pool in Cluster 3, where 86% are Dormant, before they churn completely.

Second, bundle driven merchandising over discounting. We should leverage high confidence product bundles like breakfast kits, deli lunches, and pasta nights with cross aisle displays and mailer features. These tactics deliver 15 to 30% volume lifts without the margin erosion of deep price cuts, since most top categories are inelastic with elasticities between negative 0.5 and negative 0.9.

Third, precision coupon targeting. Allocate coupons to the top 30% of households by propensity score in deciles 1 through 3, focusing on TypeA campaigns that convert 2 times better than TypeB or TypeC. Prioritize high score households with moderate to high recent spend to maximize both redemption rates and incremental revenue.

Fourth, behavior-first retention. Coupons and campaigns only extend survival when they trigger incremental trips or spend. Target low engagement households with declining trip frequency and falling recency scores with preemptive offers before they drift into the churn zone, rather than rewarding already loyal shoppers who would have visited anyway.

Together, these four pillars form a cohesive playbook that protects the high value core, reactivates dormant shoppers, grows basket size profitably, and optimizes coupon spend, all while preserving margins through smart pricing and merchandising.

## Strategic Pillar 1: Segment-First Resource Allocation

Key Findings from Customer Value Map

The household base breaks into four actionable clusters.

Cluster 2 represents Premium Loyalists with median spend of $9,700, more than 250 baskets, and recency of 2 days. This cluster has 392 households and accounts for 47% of revenue. The strategic priority is to defend and delight through concierge service, exclusive bundles, and loyalty perks.

Cluster 0 represents Growth Regulars with median spend of $3,400, weekly shopping patterns, and room to grow. This cluster has around 800 households and contributes 30 to 35% of revenue. The strategic priority is to nurture and upsell through milestone rewards, bundle trials, and trip frequency incentives.

Cluster 3 is the Reactivation Pool with declining recency and 86% Dormant status, but still reachable. This cluster has around 1,000 households and represents 15 to 20% of revenue. The strategic priority is win-back campaigns with limited time offers, price matched essentials, and multi-channel reminders.

Cluster 1 contains Lost or Churned customers with minimal activity and less than 5% revenue contribution. This cluster has around 300 households. The strategic priority is to deprioritize, with minimal spend unless reactivation tests show ROI.

Recommendations

Allocate 60% of marketing budget to Cluster 2, the Premium Loyalists. Launch a VIP program with early access to new bundles like the breakfast kits from Bundle Analysis. Offer personalized mailer features highlighting their favorite categories such as soft drinks, dairy, and proteins identified in the Price Elasticity module. Monitor for any recency decline, more than 7 days since last trip, and trigger immediate check in offers.

Invest 25% in Cluster 0, the Growth Regulars, to graduate them into Cluster 2. Design milestone journeys, for example complete 5 trips this month and earn a $10 credit. Cross sell bundles they don't yet buy. If they purchase cereal but not milk and bread, offer a breakfast bundle discount. Use coupon propensity scores to identify Cluster 0 households with high redemption likelihood and moderate spend, then target them with TypeA campaigns.

Reserve 15% for Cluster 3, the Reactivation Pool, with strict ROI gates. Launch 60 day win back campaigns for households whose recency has stretched to 30 plus days. Emphasize price matched staples like milk, bread, and eggs since Dormant households often defect on price perception. Measure incremental spend carefully. If a household redeems a coupon but doesn't increase trip frequency, as the Time to Churn module shows, the tactic failed.

Do not chase Cluster 1, the Lost or Churned segment, unless a pilot shows surprising returns. These households contribute less than 5% of revenue and rarely respond to reactivation. Better to reallocate those dollars to preventing Cluster 0 households from sliding into Cluster 3.

## Strategic Pillar 2: Bundle-Driven Merchandising Over Discounting

Key Findings from Bundle Analysis and Price Elasticity

Top product bundles that shoppers already assemble:

The Breakfast Kit is the top bundle. Cold cereal plus bread plus milk plus soft drinks or juice shows confidence from 30 to 81%, lift of 3 to 4 times, and support of 4,000 to 10,000 baskets. The action is to create a permanent breakfast endcap featuring these four categories, refreshing weekly with seasonal flavors.

The Deli Lunch Bundle follows. Lunchmeat plus bakery bread plus bag snacks shows confidence around 40%, lift of 4.2 times, and is strong during back to school and summer picnic season. The action is to co-locate the deli counter with the bread aisle and offer build your sandwich promotions.

The Pasta Night Combo is next. Dry pasta plus cheese plus milk shows confidence around 35%, lift of 4.0 times, and appeals to families, which represents 40% of Cluster 2 households with kids. The action is to bundle pasta, sauce, cheese, and salad kits in a single display and promote via weekly mailer.

Snack Pairings are popular. Baked goods plus soft drinks or bag snacks show confidence over 60%, lift of 3.5 times, and weekend and game day spikes. The action is impulse displays near checkout with party pack messaging.

Healthier Snacking is an emerging opportunity. Yogurt plus tropical fruit plus milk has lower support but high lift among health conscious shoppers. The action is to feature in wellness themed mailers and target households with low indulgent snack penetration.

Price elasticity and promotional lift:

Soft drinks, dairy, and proteins are inelastic with elasticities from negative 0.5 to negative 0.9, meaning a 10% price hike only reduces volume by 5 to 9%. We have pricing headroom to protect margins.

Displays deliver 20 to 30% volume lifts, far exceeding what we'd gain from a 5 to 10% discount.

Mailers add 10 to 20% lifts, particularly effective for dairy and select manufacturer brands.

Recommendations

Reallocate 30% of discount budget to display and mailer investments. Instead of funding blanket 15% off promotions, secure 4 to 6 permanent endcaps for the top bundles. Rotate featured bundles monthly. For example, breakfast in January and February, grilling and deli in May through July, pasta and comfort food in October and November.

Stabilize or modestly increase prices on inelastic anchors like soft drinks, milk, and bread. A 3 to 5% price increase on these categories will trim volume by only 1.5 to 4.5%, preserving margin without losing customers. Use the recovered margin to fund deeper discounts on elastic, ring building categories like premium cheeses and specialty produce that drive basket diversity.

Design cross aisle displays that pair complementary bundles. For example, place a breakfast endcap with cereal, milk, and juice near the bakery bread section so shoppers naturally add bread to the basket. Track basket penetration rates before and after display changes. Aim for 10 to 15% increases in bundle completion.

Feature bundles in weekly mailers with complete your meal messaging. Mailers reach 60 to 70% of Premium Loyalists and drive 10 to 20% lifts. Highlight bundle savings, for example $5 off when you buy all four breakfast essentials, to frame the discount as a reward for completing the mission, not a blanket cut.

## Strategic Pillar 3: Precision Coupon Targeting

Key Findings from Coupon Redemption Propensity

Propensity model performance shows an AUC around 0.75, indicating the model reliably separates redeemers from non-redeemers. The top 30% of households by score in deciles 1 through 3 capture around 68% of all redemptions and deliver 2 times lift over random targeting. TypeA campaigns convert 2 times better than TypeB or TypeC, so funding priority should follow campaign type.

Dual objective trade-offs exist. High score households in low pre-spend quartiles deliver redemption volume but minimal revenue lift. High score households in mid to high spend quartiles deliver both redemptions and meaningful incremental dollars.

Feature importance shows recent spend in the last 60 days, trip frequency, historical coupon usage, and recency are top predictors. Campaign type, TypeA versus TypeB or TypeC, significantly affects redemption rates even within the same score decile.

Recommendations

Rank all households by propensity score at the start of each quarter, then segment by campaign type and recent spend. Tier 1 is the top priority: high score TypeA households with recent spend over $200 per month. Allocate 50% of coupon budget here. Tier 2 is secondary: high score TypeA households with moderate spend between $100 and $200 per month. Allocate 25%. Tier 3 is test and learn: high score TypeB or TypeC households or low spend high scorers. Allocate 15%, measuring incremental lift strictly. Tier 4 is deprioritize: low score households regardless of spend. Allocate remaining 10% only if surplus budget exists.

Phase coupon campaigns by score decile cutoffs. In week 1, launch to decile 1 only, measuring redemption rate and campaign spend. In week 2, expand to deciles 1 and 2 if week 1 ROI meets threshold, for example $3 incremental spend per $1 coupon cost. In week 3, expand to decile 3 if cumulative ROI still positive. Stop if marginal ROI drops below breakeven and redirect remaining budget to display investments from Pillar 2.

Align coupon offers with bundle themes for cross module integration. For example, a breakfast bundle coupon offering $2 off when you buy cereal, milk, and bread, sent to high score households who already purchase cereal, will complete the basket and drive ring size. Avoid single item coupons for inelastic anchors like milk only coupons since those households would have bought milk anyway. Instead, use coupons to pull in complementary items.

Retarget non-redeemers with adjusted offers. If a high score household doesn't redeem a TypeA coupon within 14 days, send a reminder with a shorter validity window, expires in 7 days, to create urgency. For lapsed households in Cluster 3, pair coupons with we miss you messaging and price matched guarantees to overcome defection concerns.

Monitor score drift and retrain quarterly. Household behavior changes seasonally during holidays and back to school, so refresh the propensity model every quarter. Track AUC over time. Maintain above 0.73 to 0.75 before scaling campaigns. If AUC drops, revisit feature engineering, for example add product category affinity flags.

## Strategic Pillar 4: Behavior-First Retention

Key Findings from Time to Churn

Kaplan-Meier descriptive insights show coupon redeemers stay active 15 to 30 percentage points longer by week 90 than non-redeemers. Campaign reached households show similar survival advantages. Age Group 6, the older demographic cohort, churns faster than younger segments, even after adjusting for spend.

Cox proportional hazards model provides causal insights. Log trips with hazard ratio around 0.57 and log total spend with hazard ratio around 0.67 are the only significant protective factors. Coupon and campaign flags lose significance once trips and spend are controlled, meaning these tactics only extend survival when they trigger incremental behavior, more trips and higher spend, not by themselves. Age Group 6 has higher churn risk with hazard ratio over 1.0 even after controlling for behavior, suggesting demographic specific retention tactics are needed.

Recommendations

Design interventions that explicitly boost trip frequency and spend, not just redemption. The wrong approach is here's a $5 coupon, come back anytime. The household redeems once, then disappears. The right approach is complete 3 trips in the next 30 days and earn $15 in rewards. This forces incremental engagement. Tie coupon offers to multi-trip milestones so the tactic drives the behavior that actually reduces churn.

Monitor trip cadence and recency in real time, triggering preemptive offers when patterns deteriorate. Build a weekly dashboard tracking each household's days since last trip as recency, rolling 60 day trip count, and rolling 60 day spend. In the green zone for healthy households, recency is less than 7 days and trips are more than 8 in last 60 days. No intervention needed. In the yellow zone for at risk households, recency is 8 to 21 days and trips are 4 to 7 in last 60 days. Send gentle nudge messaging like we noticed you haven't visited in a while, here's a personalized offer. In the red zone for churn imminent households, recency is more than 21 days and trips are less than 4 in last 60 days. Escalate to win back campaign with time limited coupon plus we miss you messaging.

Segment retention tactics by cluster and age group. For Cluster 2, the Premium Loyalists, even a single month of declining trips is a red flag. Intervene immediately with concierge level outreach through phone call, personalized offer, or VIP event invitation. For Cluster 0, the Growth Regulars, focus on trip frequency incentives like visit 2 times per week for a month and unlock Gold status. For Cluster 3, the Reactivation Pool, these households are already in the red zone. Use aggressive win back offers, but measure incremental spend strictly. If they redeem a coupon but don't sustain trips, the tactic failed. For Age Group 6, since this cohort churns faster even after behavior adjustments, test senior specific programs like early morning shopping hours, simplified digital interfaces, or delivery partnerships.

Cross validate retention efforts with coupon propensity scores for Pillar 3 integration. Not all at risk households will respond to coupons. Prioritize those in yellow or red zones who also have high propensity scores. For example, a Cluster 0 household with declining trips but a top decile coupon score is a prime candidate for a milestone offer. A Cluster 3 household with low propensity should receive non-coupon interventions, like shop with us this week and enter to win $500.

Measure success by sustained engagement, not one time redemption. Traditional coupon ROI is incremental revenue from campaign divided by coupon cost. Retention ROI is incremental lifetime value over next 6 months divided by intervention cost. Only scale tactics that demonstrate durable trip frequency and spend increases, not just short term redemption spikes.

## Integrated Decision Framework: Putting It All Together

The four pillars are designed to work as a system, not in isolation. Here's how to orchestrate them.

Quarterly Planning Cycle

Month 1: Segmentation Refresh

Recompute RFM scores and k-means clusters to identify any households that have moved between clusters. Flag households that dropped from Cluster 2 to Cluster 0 or Cluster 0 to Cluster 3 for immediate intervention. Refresh coupon propensity scores using the most recent 90 days of transaction data.

Month 2: Bundle and Display Planning

Review association rules to confirm which bundles still show high confidence and lift. Design cross aisle displays and mailer features for the upcoming quarter, prioritizing bundles that align with seasonal demand. For example, grilling bundles in Q2, comfort food in Q4. Lock in display placements and mailer creative at least 4 weeks before launch.

Month 3: Coupon Allocation and Campaign Launch

Rank households by propensity score and segment by cluster, recent spend, and campaign type. Phase coupon rollout from decile 1 to deciles 1 and 2 to deciles 1 through 3 based on observed ROI. Tie coupon offers to bundle themes, for example $3 off breakfast bundle versus generic $3 off any purchase.

Ongoing: Real-Time Churn Monitoring

Run weekly recency and trip frequency checks using the dashboard. Trigger preemptive offers for yellow zone households, escalate to win back campaigns for red zone households. Measure retention ROI monthly. If an intervention doesn't sustain trips, kill it and redirect budget to display investments in Pillar 2.

Resource Allocation Blueprint

Here's an example using a $100,000 quarterly marketing budget.

VIP and Concierge Program for Cluster 2 retention gets $35,000. Target the 392 Premium Loyalists. Expected outcome is maintaining 47% revenue share and preventing any defections.

Cross Aisle Displays and Endcaps for the Bundle Pillar gets $25,000. Target all shoppers with emphasis on Clusters 2 and 0. Expected outcome is 15 to 30% volume lift on featured bundles and growing ring size by 10%.

Weekly Mailer Features for the Bundle Pillar gets $15,000. Target 60 to 70% of base with Cluster 2 and 0 priority. Expected outcome is 10 to 20% lift on mailer featured categories.

Precision Coupon Campaigns for the Coupon Pillar gets $15,000. Target top 30% by propensity score with TypeA focus. Expected outcome is 2 times lift versus random targeting and capturing 68% of redemptions.

Preemptive Retention Offers for the Churn Pillar gets $7,000. Target yellow zone households in Clusters 0 and 2. Expected outcome is preventing 50 plus households from sliding into red zone.

Win-Back Campaigns for the Churn Pillar gets $3,000. Target red zone households in Cluster 3. Expected outcome is reactivating 5 to 10% of Dormant households with ROI gated strictly.

Success Metrics by Pillar

For Segmentation, the primary KPI is Cluster 2 revenue share. Target is maintaining at or above 47%. Review monthly.

For Bundles and Merchandising, the primary KPI is bundle penetration rate, the percentage of baskets containing at least 3 bundle items. Target is increasing from baseline 22% to 30%. Review quarterly.

For Coupon Targeting, the primary KPI is redemption rate in top 3 deciles. Target is at or above 18% versus 9% baseline. Review per campaign.

For Retention, the primary KPI is percentage of yellow zone households that recover to green. Target is at or above 60%. Review via weekly dashboard.

Continuous Improvement Loop

Test and learn. Every quarter, reserve 10% of budget for experimental tactics. For example, senior specific bundles for Age Group 6 or digital coupon delivery for tech savvy households.

Score recalibration. Retrain coupon propensity and RFM clustering models quarterly to capture evolving behavior.

Cross-module audits. Every 6 months, audit whether tactics from one pillar are undermining another. For example, are deep discounts eroding the price discipline from Pillar 2?

Executive reviews. Present integrated results, not siloed by module, to leadership quarterly, emphasizing total revenue impact and margin preservation.

## Final Recommendations: The 90-Day Action Plan

To translate these strategic pillars into immediate action, we recommend the following 90 day roadmap.

Days 1 to 30: Foundation and Quick Wins

Launch VIP retention program for Cluster 2, the Premium Loyalists. Send personalized thank you for being a valued customer mailers with exclusive early access to new bundles. Assign account managers to the top 100 households by lifetime value for proactive check ins.

Install 2 permanent endcaps for top bundles. The breakfast kit includes cereal, milk, bread, and juice. The deli lunch includes meats, bread, and snacks. Measure basket penetration weekly. Target 25% of shoppers picking up at least 3 of the 4 bundle items.

Run pilot coupon campaign on top propensity decile with TypeA only. Offer $5 off when you complete a breakfast bundle. Send to 200 households in decile 1, measure redemption rate and campaign spend after 14 days. If redemption exceeds 20% and incremental spend exceeds $50 per household, expand to deciles 1 and 2 in Month 2.

Deploy churn dashboard for real time monitoring. Integrate recency, trip count, and spend into a weekly email alert for marketing ops. Flag any Cluster 2 household whose recency exceeds 7 days for immediate outreach.

Days 31 to 60: Scale and Optimize

Expand display program to 4 to 6 endcaps. Add pasta night, snack pairings, and healthier snacking bundles. Coordinate with weekly mailer team to feature the same bundles for cross channel reinforcement.

Broaden coupon targeting to deciles 1 through 3 based on pilot results. Maintain TypeA priority, add top scoring TypeB households if budget allows. Shift messaging from generic $5 off to bundle specific messaging like complete your pasta night and save.

Launch preemptive retention offers for yellow zone households. We noticed you haven't visited in 10 days, here's a personalized offer to welcome you back. Measure trip recovery rate, the percentage returning to green zone within 30 days.

Stabilize prices on inelastic anchors like soft drinks, milk, and bread. Implement 3% price increase, monitor volume closely. Use recovered margin to fund deeper discounts on ring building categories like specialty cheeses and organic produce.

Days 61 to 90: Measure, Learn, Iterate

Conduct mid quarter performance review. Compare actual versus target on all KPIs: Cluster 2 revenue share, bundle penetration, coupon redemption, and churn recovery. Identify underperforming tactics and reallocate budget to winners.

Refresh segmentation and propensity scores. Recompute RFM and k-means clusters to catch any households that shifted between clusters. Retrain coupon propensity model on the most recent 90 days of data.

Design Q2 test and learn experiments. For example, test senior specific bundles with easy open packaging and smaller sizes for Age Group 6 to address their higher churn risk. Another example is piloting digital only coupons for younger households to reduce print and postage costs.

Present integrated results to leadership. Show total revenue impact not siloed by module. For example, Cluster 2 revenue share held at 47% despite market pressures, bundle displays added X incremental revenue, coupon ROI was Y per dollar spent. Secure budget approval for Q2 expansion based on demonstrated ROI.

The four strategic pillars, segment first allocation, bundle driven merchandising, precision coupon targeting, and behavior first retention, are not theoretical frameworks but practical playbooks grounded in rigorous analysis.

Together, these five modules form an integrated system where each pillar reinforces the others. Segmentation feeds targeting, bundles inform promotions, pricing protects margins, coupons drive selective volume, and retention focuses on sustainable engagement.
